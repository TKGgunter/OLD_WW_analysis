````python
def bin( arr, range=(-1.25, 1.25), bins=50, norm=False ):

    binned = np.histogram(arr, range=range, bins=bins)[0].astype(float)
    if norm == True:
        return binned / binned.sum()
    else:
        return binned
````

````python
#Example 
#It works!!!
mid_points = [-.1, 0.2]
size = [10000, 5000]
scale = [.5, .25]

#bins = np.histogram(np.random.normal(loc=mid_points[0], scale=scale[0]))[1]

gauss0 = np.random.normal(loc=mid_points[0], scale=scale[0], size=size[0])
gauss1 = np.random.normal(loc=mid_points[1], scale=scale[1], size=size[1])

data = bin(gauss0) + bin(gauss1)
data = data / data.sum()

mc_g0 = bin(gauss0, norm=True)
mc_g1 = bin(gauss1, norm=True)

mc = mc_g0 + mc_g1
def func( n_bin, a, b):
    return a * mc_g0 + b * mc_g1
    #return a * mc_g0[n_bin] + b * mc_g1[n_bin]

def min_func(x):
    fit = func(0, x[0], x[1])
    #print x, (np.abs(fit - data)**2. / data).sum()
    return (np.abs(fit - data)**2. / data).sum()

def func2( n_bin, a):
    return a * mc_g0 + (1-a) * mc_g1

temp = curve_fit(func, [i for i in range(data.shape[0])], data)
temp1 = minimize(min_func, [1.,1.], method='SLSQP')
temp2 = curve_fit(func2, [i for i in range(data.shape[0])], data)
print "From fit: ", temp[0]
print "From fit1: ", temp1["x"]
print "From fit2: ", temp2[0], 1 - temp2[0]
print "Expected: ",2/3., 1./3. 
````
