# Comparing the nTuples of Sept and Pre-March 

Oct. 7th

I found that there was a large discrepancy in the measured WZ, ZZ, and later DY and TTbar process during random forest optimization.  This occured after adding Z to tau tau identification and lepton charge information. After investigating it was found that the WZ and ZZ processes required different cross-sections and all total number of events.  DY and TTbar samples did not require scaling updates. In validating the updated scaling it was determine that the total WW yeilds were significatly smaller compared to previous ntuple samples about 2x. 


Additional work was done to compare MC to data.  Initial hypothesis were the current WW sample was under scaled, but after comparing results with Nate Odell it was concluded that there was a factor 2 too much data.

Note: ntuple alterations allowed for 3 WW files to the previous 2.
